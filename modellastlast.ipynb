```python
import pandas as pd
from pycaret.time_series import *
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
import matplotlib.pyplot as plt

# กำหนดค่ามาตรฐานคุณภาพอากาศ
AIR_QUALITY_STANDARDS = {
    'GOOD': 15,
    'MODERATE': 25,
    'UNHEALTHY': 37.5,
    'VERY_UNHEALTHY': 75,
}

# ฟังก์ชันสำหรับจัดประเภทคุณภาพอากาศตามค่า PM2.5
def classify_air_quality(pm25_value):
    if pm25_value <= AIR_QUALITY_STANDARDS['GOOD']:
        return "อากาศดีมาก! 🎉"
    elif pm25_value <= AIR_QUALITY_STANDARDS['MODERATE']:
        return "อากาศดี 😊"
    elif pm25_value <= AIR_QUALITY_STANDARDS['UNHEALTHY']:
        return "อากาศไม่ดี 😷"
    elif pm25_value <= AIR_QUALITY_STANDARDS['VERY_UNHEALTHY']:
        return "อากาศแย่มาก! ⚠️"
    else:
        return "อากาศอันตราย! 🚨"

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv('cleaned_data_withtime.csv')

# แปลงคอลัมน์ timestamp เป็น datetime และตั้งค่าเป็น index
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%m/%d/%Y %H:%M')
df.set_index('timestamp', inplace=True)

# ตรวจสอบข้อมูลที่ซ้ำกันในคอลัมน์ timestamp
if df.index.duplicated().any():
    print("พบข้อมูลที่ซ้ำกันในคอลัมน์ timestamp")
    df = df[~df.index.duplicated(keep='first')]  # ลบข้อมูลที่ซ้ำกัน โดยเก็บค่าแรกไว้

# กำหนดค่า freq ให้กับ index
freq = pd.infer_freq(df.index)
if freq is None:
    freq = 'D'  # หากไม่สามารถระบุความถี่ได้ ให้กำหนดเป็นรายวัน (daily)
    df = df.asfreq(freq)  # กำหนดความถี่ให้กับข้อมูล

# เติมค่าที่หายไปด้วย forward fill
df = df.ffill()

# สร้าง lag ข้อมูล 1-7 วัน
for i in range(1, 8):
    df[f'pm_2_5_lag_{i}'] = df['pm_2_5'].shift(i)
    df[f'temperature_lag_{i}'] = df['temperature'].shift(i)
    df[f'humidity_lag_{i}'] = df['humidity'].shift(i)

# ลบแถวที่มีค่า NaN ทิ้ง
df_cleaned = df.dropna()

# แบ่งข้อมูลเป็นชุดฝึกและชุดทดสอบ
train_df = df_cleaned.iloc[:-7]
test_df = df_cleaned.iloc[-7:]

# Features และ Target
features = [f'pm_2_5_lag_{i}' for i in range(1, 8)] + \
           [f'temperature_lag_{i}' for i in range(1, 8)] + \
           [f'humidity_lag_{i}' for i in range(1, 8)]

target = 'pm_2_5'

# Setup ข้อมูลด้วย PyCaret
# หมายเหตุ: ไม่ต้องส่ง index เข้าไปใน setup() โดยตรง
setup(
    data=train_df,
    target=target,
    session_id=123,
    fh=7,
    fold=3,
    numeric_imputation_target='ffill',  # เติมค่าที่หายไปด้วย forward fill
    numeric_imputation_exogenous='ffill',  # เติมค่าที่หายไปด้วย forward fill
)

# สร้างและ Finalize โมเดล
arima_model = create_model('arima', order=(1, 1, 1), seasonal_order=(1, 1, 1, 7))# ARIMA(p, d, q) SARIMA(P, D, Q, s)
arima_model = finalize_model(arima_model)

# ทำนายผลลัพธ์
# แปลง index ของ test_df ให้เป็น PeriodIndex เพื่อให้ตรงกับรูปแบบที่ PyCaret ใช้
test_df.index = test_df.index.to_period(freq)

# ทำนายค่า PM2.5
predictions = predict_model(arima_model, fh=7, X=test_df.drop(columns=target))

# เพิ่มคอลัมน์คุณภาพอากาศ
test_df['Predicted_PM2.5'] = predictions
test_df['Air_Quality'] = test_df['Predicted_PM2.5'].apply(classify_air_quality)

# แสดงผลลัพธ์
print(test_df[['Predicted_PM2.5', 'Air_Quality']])

# คำนวณค่าประสิทธิภาพ
mae = mean_absolute_error(test_df[target], predictions) # ค่าความผิดพลาดเฉลี่ยแบบสัมบูรณ์
mape = mean_absolute_percentage_error(test_df[target], predictions) * 100 #ค่าความผิดพลาดเฉลี่ยแบบเปอร์เซ็นต์ 
r2 = r2_score(test_df[target], predictions) #ค่าความสามารถของโมเดลในการอธิบายความแปรปรวนของข้อมูล 

print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape:.2f}%")
print(f"R2: {r2:.2f}")

# พล็อตกราฟเปรียบเทียบค่าจริงกับค่าทำนาย
plt.figure(figsize=(12, 6))
plt.plot(test_df.index.to_timestamp(), test_df[target], label='Actual', color='blue', marker='o')
plt.plot(test_df.index.to_timestamp(), predictions, label='Predicted', color='red', marker='o')
plt.title('Actual vs Predicted PM 2.5', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('PM 2.5', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
